# Book

```python
import numpy as np
import pandas as pd
```

```python
dataset_book = pd.read_csv('bookAR (1).csv',sep=',')
dataset_book = dataset_book.replace(np.nan, '', regex=True)
dataset_book
```

```python
dataset_book
transactions = []
for i in range(0, 2000):
    transactions.append([str(dataset_book.values[i,u]) for u in range(0, 11)]) 
```

```python
book_series  = pd.DataFrame(pd.Series(transactions))
book_series 
book_series.columns = ["business"]
book_series
```

```python
X = dataset_book
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets_book = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets_book
```

```python
frequent_itemsets_book.sort_values('support',ascending = False,inplace=True)
frequent_itemsets_book.sort_values
```

```python
book_rules = association_rules(frequent_itemsets_book, metric="lift", min_threshold=1)
book_rules.head(20)
book_rules.sort_values('lift',ascending = False).head(10)
```

# Groceries

```python
import numpy as np
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dataset = pd.read_csv('groceries - groceries.csv',sep=',')
dataset = dataset.drop(['Item(s)'], axis=1)
dataset = dataset.replace(np.nan, '', regex=True)
dataset 
```

```python
dataset
transactions = []
for i in range(0, 9835):
    transactions.append([str(dataset.values[i,u]) for u in range(1, 31)]) 
```

```python
groceries_series  = pd.DataFrame(pd.Series(transactions))
groceries_series 
groceries_series.columns = ["business"]
groceries_series
```

```python
X1 = groceries_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X1
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X1, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets 
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)
```
